```python
import cv2

import numpy as np
```

```python
def rectContours(contours):
    
    rectCont=[]
    for i in contours:
        area=cv2.contourArea(i) #find all area in contours 
        #print("Area",area)
        if area>50:
            par=cv2.arcLength(i,True)# TRUE=> colsed one           #total lenght of this contours
            approx=cv2.approxPolyDP(i,0.02*par,True) #hpw many corner points it has   #.02 => resolution it can be changed based on our par(total lenght)
           # print("Corner Points",len(approx))  #**without **len** they give us actual points(coordinates i think) not how many points we have
            if len(approx)==4:
                rectCont.append(i)
                
    rectCont=sorted(rectCont,key=cv2.contourArea,reverse=True) #sort decending  
    return rectCont



def getCornerPoints(cont):
     par=cv2.arcLength(cont,True)
     approx=cv2.approxPolyDP(cont,0.02*par,True)
     return approx
    
    
    
    
def reorder(myPoints):
    myPoints=myPoints.reshape(4,2)
    myPointsNew=np.zeros((4,1,2),np.int32)
    add=myPoints.sum(1)
    
    #print(myPoints)
    #print(add)
    myPointsNew[0]=myPoints[np.argmin(add)] #[0,0]
    myPointsNew[3]=myPoints[np.argmax(add)] #[w,h]
    
    diff=np.diff(myPoints,axis=1) #to Knoe which one comes first
    myPointsNew[1]=myPoints[np.argmin(diff)] #[w,0]
    myPointsNew[2]=myPoints[np.argmax(diff)] #[0,h]
    #print(diff)
    
    return myPointsNew



def splitBoxes(img):
    rows=np.vsplit(img,5)
    
    boxes=[]
    for r in rows:
        cols=np.hsplit(r,4)
        for box in cols:
            boxes.append(box)
            #cv2.imshow("split",box)
    return boxes        
        
    
    
def showAnswers(img,myIndex,grading,corAns,questions,answers):
    secW=int(img.shape[0]/answers)
    secH=int(img.shape[1]/questions)
    
    for x in range(0,questions):
        myAns=myIndex[x]
        #find the center value of our given box 
        cX=(myAns*secW)+(secW//2)
        cY=(x*secH)+(secH//2)
        
        if grading[x]==1:
            myColor=(0,255,0)
        else:
            myColor=(0,0,255)
            correctAns=corAns[x]
            cv2.circle(img,((correctAns*secW)+secW//2,(x*secH)+secH//2),30,(0,255,0),cv2.FILLED)
            
               
        cv2.circle(img,(cX,cY),40,myColor,cv2.FILLED)
            
        
    return img
```

```python
path="multiple choice.jpg"
width=500
height=500
questions=5
answers=4
corAns=[0,1,1,2,3]



img = cv2.imread(path)
img=cv2.resize(img,(width,height))
imgFinal=img.copy()
imgContours=img.copy()
imgBiggestContours=img.copy()

imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
imgBlur=cv2.GaussianBlur(imgGray,(5,5),1)
imgCanny=cv2.Canny(imgBlur,10,50)

#Finding all contours
contours,hierarchy=cv2.findContours(imgCanny,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
cv2.drawContours(imgContours,contours,-1,(0,255,0),4)
 
#find rectangls
rectCont=rectContours(contours)
biggestContour=getCornerPoints(rectCont[0])
#print(biggestContour)
#print(biggestContour.shape) #what kinds of pointes that we have
gradeRect=getCornerPoints(rectCont[2])

if biggestContour.size != 0 and gradeRect.size != 0:
    cv2.drawContours(imgBiggestContours,biggestContour,-1,(0,255,0),10)
    cv2.drawContours(imgBiggestContours,gradeRect,-1,(255,0,0),10)
    
    biggestContour=reorder(biggestContour)
    gradeRect=reorder(gradeRect)
    
    pt1=np.float32(biggestContour)
    pt2=np.float32(([0,0],[width,0],[0,height],[width,height]))
    matrix=cv2.getPerspectiveTransform(pt1,pt2) #our transformation matrix
    imgWarpColored=cv2.warpPerspective(img,matrix,(width,height))
    
    ptG1=np.float32(gradeRect)
    ptG2=np.float32(([0,0],[350,0],[0,150],[350,150]))
    matrixG=cv2.getPerspectiveTransform(ptG1,ptG2) #our transformation matrix
    imgWarpGrade=cv2.warpPerspective(img,matrixG,(350,150))
    
    imgWarpGray=cv2.cvtColor(imgWarpColored,cv2.COLOR_BGR2GRAY)
    imgThresh=cv2.threshold(imgWarpGray,150,255,cv2.THRESH_BINARY_INV)[1]
    
    boxes=splitBoxes(imgThresh)
    
    # getting non zero pixel value of each box
    myPixelVal=np.zeros((questions,answers)) #array 5*4
    countC=0
    countR=0
    
    for image in boxes:
        totalPixels=cv2.countNonZero(image)
        myPixelVal[countR][countC]=totalPixels
        countC +=1
        if (countC == answers): countR +=1 ; countC=0
    #print(myPixelVal)
    
    #finding index value of the marking
    myIndex=[]
    for x in range(0,questions):
        arr=myPixelVal[x]
        #print("arr",arr)
        myIndexVal=np.where(arr==np.amax(arr))
        #print(myIndexVal[0])
        myIndex.append(myIndexVal[0][0]) #saving to the list while easy to compare
    #print(myIndex)
    
    
    #grading

    grading=[]
    for x in range(0,questions):
        if corAns[x]==myIndex[x]:
            grading.append(1)
        else:
            grading.append(0)
            #print(grading)
    
    #final garde
    score=(sum(grading)/(questions))*100
    #print(score)
    
    #displaying answers
    imgResult=imgWarpColored.copy()
    imgResult=showAnswers(imgResult,myIndex,grading,corAns,questions,answers)
    
    imgRawDrawing=np.zeros_like(imgWarpColored)
    imgRawDrawing=showAnswers(imgRawDrawing,myIndex,grading,corAns,questions,answers)
    invMatrix=cv2.getPerspectiveTransform(pt2,pt1) #our transformation matrix(inverse matrix)
    imgInvWarp=cv2.warpPerspective(imgRawDrawing,invMatrix,(width,height))
    

    
    imgRawGrade=np.zeros_like(imgWarpGrade)
    cv2.putText(imgRawGrade,str(int(score))+"%",(85,100),cv2.FONT_HERSHEY_COMPLEX,3,(0,0,255),10) # 3=>scale , 10=>thickness
    invMatrixG=cv2.getPerspectiveTransform(ptG2,ptG1) #our transformation matrix
    imgInvGrade=cv2.warpPerspective(imgRawGrade,invMatrixG,(width,height))
    

    
    
    imgFinal=cv2.addWeighted(imgFinal,1,imgInvWarp,1,0)
    imgFinal=cv2.addWeighted(imgFinal,1,imgInvGrade,1,0)
    
    
    
    
    
    



#cv2.imshow("img",img)
#cv2.imshow("imgRsize",img)
#cv2.imshow("imgGray",imgGray)
#cv2.imshow("imgBlur",imgBlur)

#cv2.imshow("imgContours",imgContours)
#cv2.imshow("imgBiggestContours",imgBiggestContours)
#cv2.imshow("imgWarpColored",imgWarpColored)
#cv2.imshow("imgWarpGrade",imgWarpGrade)
#cv2.imshow("imgThresh",imgThresh)
#cv2.imshow("boxes",boxes[5])
#cv2.imshow("imgResult",imgResult)
#cv2.imshow("imgRawDrawing",imgRawDrawing)
#cv2.imshow("imgInvWarp",imgInvWarp)
#cv2.imshow("imgFinal",imgFinal)
#cv2.imshow("imgInvGrade",imgInvGrade)
cv2.imshow("Ggrade",imgFinal)










#cv2.imshow("Canny",imgCanny)
cv2.waitKey(0)
```
